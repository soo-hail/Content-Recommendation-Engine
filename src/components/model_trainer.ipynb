```python
import pickle
```

```python
with open('artifacts/pivoted_df.pkl', 'rb') as f:
        pivoted_df = pickle.load(f)
```

```python
pivoted_df.head()
```

```python
corrMatrix = pivoted_df.corr(method='pearson')
```

```python
corrMatrix.head(50)
```

```python
def get_similar_movie(movie_id, rating):
    similar_ratings = corrMatrix[movie_id] * (rating - 2.5)
    similar_ratings = similar_ratings.sort_values(ascending = False)
    
    return similar_ratings
```

```python
get_similar_movie(377, 1)
```

```python
import pandas as pd
user_rating = [(322, 4), (12107, 1), (3989, 3), (769, 2), (11, 3)]

similar_scores = []
for movie, rating in user_rating:
    similar_scores.append(get_similar_movie(movie, rating))

index = [f'movie_{id}' for id,_ in user_rating]

similar_scores_df = pd.DataFrame(similar_scores, index = index) # CONVERT LIST TO A DATA-FRAME.
similar_scores_df.head(10) 
```

```python
# SUM ALL THE CORRELATIONS(SIMILARITY) AND GET TOP 10 MOVIES.
print('TOP RECOMENDATIONS: ')
similar_scores_df.sum().sort_values(ascending=False).head(10)
```
